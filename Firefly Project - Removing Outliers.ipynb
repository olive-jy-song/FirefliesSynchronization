```python
import numpy as np
import matplotlib.pyplot as plt 
N = 100
pi = np.pi 
w = 50
dt = 0.01
Kr = 0.4
step_size = 200 
flies_index = [i for i in range(N)]
flies_ini_phase = np.random.uniform(-pi,pi,N) 
curr_phase = flies_ini_phase 
phases_minus_mean = [] 
```

```python
def calculate_phases(curr_phase, flies_index ,step_size, dt): 
    j = 0
    for t in range (0,step_size):    
        phase_mean = np.mean(curr_phase)  
        for i in range (0,len(curr_phase)): 
            curr_phase[i] += w + dt * Kr * np.sin(phase_mean-curr_phase[i]) 
    phases_minus_mean = curr_phase - np.mean(curr_phase) 
    while j < len(curr_phase): 
        if phases_minus_mean[j] > pi or phases_minus_mean[j] < -pi: 
            curr_phase = np.delete(curr_phase,j,None)
            flies_index = np.delete(flies_index,j,None) 
            phases_minus_mean = np.delete(phases_minus_mean,j,None)
        j+=1 
    return curr_phase, flies_index, phases_minus_mean 

def plot_phases(curr_phase): 
    plt.scatter(flies_index, curr_phase) 
    plt.xlabel('Index of Fireflies')
    plt.ylabel('Phases') 
```

```python
plot_phases(curr_phase) 
plt.title('Randomly Initialized Fireflies Phases')
print("The mean phase is ", np.mean(curr_phase))
```

```python
curr_phase, flies_index, phases_minus_mean = calculate_phases(curr_phase, flies_index, step_size, dt) 
print(len(phases_minus_mean))
print(len(curr_phase))
print(len(flies_index))
plot_phases(phases_minus_mean) 
```
